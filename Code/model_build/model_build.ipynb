# 数据处理

## 站点分布

```python
route = list(actual_sequences.keys())[22]
stops = route_data[route]["stops"]

latitudes = [stop["lat"] for stop in stops.values()]
longitudes = [stop["lng"] for stop in stops.values()]

plt.figure(figsize=(8, 6))
plt.scatter(longitudes, latitudes, c='blue', marker='o')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Stops Map')
plt.grid(True)
plt.show()
```
