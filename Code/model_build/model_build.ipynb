# 数据处理

## 站点分布及快递员实际路线

```python
def plot_route(route_num, outlier=''):
    route = list(actual_sequences.keys())[route_num]
    stops = route_data[route]["stops"]
    n = len(stops)
    score = route_data[route]["route_score"]
    
    stops_in_order = [None for _ in range(n)]
    for stop, num in actual_sequences[route]["actual"].items():
        stops_in_order[num] = stop
    
    latitudes = [stops[stop]["lat"] for stop in stops_in_order if stop != outlier]
    longitudes = [stops[stop]["lng"] for stop in stops_in_order if stop != outlier]
    
    plt.figure(figsize=(8, 6))
    plt.scatter(longitudes, latitudes, c='blue', marker='o')
    plt.plot(longitudes, latitudes, c='red')
    
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'{score} Score Stops Map')
    plt.grid(True)
    plt.show()
```

### 低分数

```python
plot_route(31)
```

### 中等分数

```python
plot_route(22)
```

### 高分数

#### 全景

```python
plot_route(23)
```

#### 略缩图

```python
plot_route(23, "YI")
```
